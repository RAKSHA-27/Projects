```python
%%writefile bird_app.py
import streamlit as st
from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.capuchin_bird_report import app as capuchin_bird_report_app
from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.sapsucker_bird_report import app as sapsucker_bird_report_app
from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.junco_bird_report import app as junco_bird_report_app
from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.fishcrow_bird_report import app as fishcrow_bird_report_app
from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.magpie_bird_report import app as magpie_bird_report_app

from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.capuchin_bird_counter import count_capuchin_birds
from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.sapsucker_bird_counter import count_sapsucker_birds
from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.junco_bird_counter import count_junco_birds
from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.fishcrow_bird_counter import count_fishcrow_birds
from drive.MyDrive.FYP_Datasets.Py_files.Counting_py_files.magpie_bird_counter import count_magpie_birds

from drive.MyDrive.FYP_Datasets.Py_files.final_classify.classify_app import load_model_and_predict
from drive.MyDrive.FYP_Datasets.Py_files.final_classify.rec_app import recommend
import tempfile

from drive.MyDrive.FYP_Datasets.Py_files.search_app import search

from pathlib import Path
import base64

# Set page title
st.set_page_config(page_title="Aviatory Club", page_icon=":blue_circle:")

# Get the directory path of the script
script_dir = Path(__file__).parent

# Construct the absolute path to the image file
image_path = str(script_dir / "drive/MyDrive/FYP_Datasets/UI/wallpaper.jpg")

# Set the background image
def set_bg_hack(main_bg):
    """ A function to unpack an image from root folder and set as bg.
    Returns
    -------
    The background.
    """
    # set bg name
    main_bg_ext = "jpg"
    st.markdown(
        f"""
        <style>
        .stApp {{
            background-image: url("data:image/{main_bg_ext};base64,{base64.b64encode(open(main_bg, "rb").read()).decode()}");
            background-size: 100% calc(100% - 0px);
            /* Adjust the value '0px' as per your taskbar height */
            background-position: center;
            background-repeat: no-repeat;
            background-attachment: fixed;
            padding-bottom: 0px;
            /* Adjust the value as per your taskbar height */
        }}
        </style>
        """,
        unsafe_allow_html=True,
    )

set_bg_hack(image_path)

# Set button color to 322C2B
st.markdown(
    """
    <style>
    div[data-testid="stButton"] > button {
        background-color: #322C2B;
    }
    </style>
    """,
    unsafe_allow_html=True,
)

# Set button text color to white
st.markdown(
    """
    <style>
    div[data-testid="stButton"] > button {
        color: white;
    }
    </style>
    """,
    unsafe_allow_html=True,
)

def main():
    menu = ["Home", "Classify", "Monitor"]
    choice = st.sidebar.selectbox("Menu", menu)

    if choice == "Home":
        title = "Bird Classification & Monitoring"
        st.title(title)

        st.subheader("Spot, Record, and Identify: Your Birdwatching Buddy! Listen, Love and Learn about the birds and their effect on the Nature.")

        st.markdown(
        """
        <style>
        .main .block-container {
            background-color: #F5E8DD;
            border-radius: 20px;
            padding: 20px;
            text-align: center;
            margin: 100px; /* Adjust margin as needed */
        }
        </style>
        """,
        unsafe_allow_html=True
        )

        search()

    elif choice == "Classify":
        title = "Bird Call Classification"
        st.title(title)

        st.markdown(
        """
        <style>
        .main .block-container {
            background-color: #CCD3CA;
            border-radius: 20px;
            padding: 20px;
            text-align: center;
            margin: 100px; /* Adjust margin as needed */
        }
        </style>
        """,
        unsafe_allow_html=True
        )

        page_1()
    elif choice == "Monitor":
        title = "Bird Call Monitoring"
        st.title(title)

        st.markdown(
        """
        <style>
        .main .block-container {
            background-color: #D6DAC8;
            border-radius: 20px;
            padding: 20px;
            text-align: center;
            margin: 100px; /* Adjust margin as needed */
        }
        </style>
        """,
        unsafe_allow_html=True
        )

        page_2()

def page_1():

    st.markdown("### Upload an audio file")
    uploaded_file = st.file_uploader("", type=["mp3", "wav", "ogg"])

    # Display file path if a file is uploaded
    if uploaded_file is not None:
        st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

        with tempfile.NamedTemporaryFile(delete=False) as temp_file:
            temp_file_path = temp_file.name
            temp_file.write(uploaded_file.read())

        real_name = load_model_and_predict(temp_file_path)

        st.markdown(
            f"""
            <div style="text-align: center;">
                <span style="font-size: 1.5rem; font-weight: bold;">
                    The bird classified is:
                </span>
                <br>
                <div style="width: 210px; height: 120px; background-color: #BBC3A4; display: flex; justify-content: center; align-items: center; margin: 0 auto;border-radius: 20px;">
                    <span style="font-size: 2rem;">{real_name}</span>
                </div>
                <br>
            </div>
            """,
            unsafe_allow_html=True
        )

        recommend(real_name)

def page_2():

    count_menu = ["No Selection","Capuchin Bird", "Yellow Bellied Sapsucker", "Dark Eyed Junco","Fishcrow", "Black Billed Magpie"]
    count_choice = st.selectbox("Menu", count_menu)

    detect_button_clicked = False

    if count_choice == "Capuchin Bird":
        st.markdown("## Capuchin")

        #st.markdown("#### Enter the path to the audio file")
        #audio_file_path = st.text_input("File Path")
        #NEWLY ADDED STUFF
        st.markdown("### Upload an audio file")
        uploaded_file = st.file_uploader("", type=["mp3", "wav", "ogg"])
        #NEWLY ADDED STUFF ENDS HERE

        detect_button = st.button("Detect")

        st.session_state.disabled = True

        if detect_button and uploaded_file: #change uploaded_file to audio_file_path
            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            calls= count_capuchin_birds(temp_file_path) #change it to audio_file_path
            st.markdown(
                f"""
                <div style="text-align: center;">
                    <span style="font-size: 1.5rem; font-weight: bold;">
                        The number of Capuchin Bird calls detected is:
                    </span>
                    <br>
                    <div style="width: 80px; height: 80px; background-color: #B5C18E; border-radius: 50%; display: flex; justify-content: center; align-items: center; margin: 0 auto;">
                        <span style="font-size: 1.5rem;">{calls}</span>
                    </div>
                    <br>
                </div>
                """,
                unsafe_allow_html=True
            )
            st.session_state.disabled = False

        get_report = st.button("Get Report", disabled=st.session_state.disabled)

        if get_report:

            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            capuchin_bird_report_app(temp_file_path)

    elif count_choice == "Yellow Bellied Sapsucker":
        st.markdown("## Yellow Bellied Sapsucker")

        #st.markdown("#### Enter the path to the audio file")
        #audio_file_path = st.text_input("File Path")

        #NEWLY ADDED STUFF
        st.markdown("### Upload an audio file")
        uploaded_file = st.file_uploader("", type=["mp3", "wav", "ogg"])
        #NEWLY ADDED STUFF ENDS HERE

        detect_button = st.button("Detect")

        st.session_state.disabled = True

        if detect_button and uploaded_file: #change uploaded_file to audio_file_path
            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            calls= count_sapsucker_birds(temp_file_path) #change it to audio_file_path
            st.markdown(
                f"""
                <div style="text-align: center;">
                    <span style="font-size: 1.5rem; font-weight: bold;">
                        The number of Yellow Bellied Sapsucker Bird calls detected is:
                    </span>
                    <br>
                    <div style="width: 80px; height: 80px; background-color: #B5C18E; border-radius: 50%; display: flex; justify-content: center; align-items: center; margin: 0 auto;">
                        <span style="font-size: 1.5rem;">{calls}</span>
                    </div>
                    <br>
                </div>
                """,
                unsafe_allow_html=True
            )
            st.session_state.disabled = False

        get_report = st.button("Get Report", disabled=st.session_state.disabled)

        if get_report:

            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            sapsucker_bird_report_app(temp_file_path)

    elif count_choice == "Dark Eyed Junco":
        st.markdown("## Dark Eyed Junco")

        #st.markdown("#### Enter the path to the audio file")
        #audio_file_path = st.text_input("File Path")

        #NEWLY ADDED STUFF
        st.markdown("### Upload an audio file")
        uploaded_file = st.file_uploader("", type=["mp3", "wav", "ogg"])
        #NEWLY ADDED STUFF ENDS HERE

        detect_button = st.button("Detect")

        st.session_state.disabled = True

        if detect_button and uploaded_file: #change uploaded_file to audio_file_path
            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            calls= count_junco_birds(temp_file_path) #change it to audio_file_path
            st.markdown(
                f"""
                <div style="text-align: center;">
                    <span style="font-size: 1.5rem; font-weight: bold;">
                        The number of Dark Eyed Junco Bird calls detected is:
                    </span>
                    <br>
                    <div style="width: 80px; height: 80px; background-color: #B5C18E; border-radius: 50%; display: flex; justify-content: center; align-items: center; margin: 0 auto;">
                        <span style="font-size: 1.5rem;">{calls}</span>
                    </div>
                    <br>
                </div>
                """,
                unsafe_allow_html=True
            )
            st.session_state.disabled = False

        get_report = st.button("Get Report", disabled=st.session_state.disabled)

        if get_report:

            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            junco_bird_report_app(temp_file_path)

    elif count_choice == "Fishcrow":
        st.markdown("## Fishcrow")

        #st.markdown("#### Enter the path to the audio file")
        #audio_file_path = st.text_input("File Path")

        #NEWLY ADDED STUFF
        st.markdown("### Upload an audio file")
        uploaded_file = st.file_uploader("", type=["mp3", "wav", "ogg"])
        #NEWLY ADDED STUFF ENDS HERE

        detect_button = st.button("Detect")

        st.session_state.disabled = True

        if detect_button and uploaded_file: #change uploaded_file to audio_file_path
            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            calls= count_fishcrow_birds(temp_file_path) #change it to audio_file_path
            st.markdown(
                f"""
                <div style="text-align: center;">
                    <span style="font-size: 1.5rem; font-weight: bold;">
                        The number of Fishcrowe Bird calls detected is:
                    </span>
                    <br>
                    <div style="width: 80px; height: 80px; background-color: #B5C18E; border-radius: 50%; display: flex; justify-content: center; align-items: center; margin: 0 auto;">
                        <span style="font-size: 1.5rem;">{calls}</span>
                    </div>
                    <br>
                </div>
                """,
                unsafe_allow_html=True
            )
            st.session_state.disabled = False

        get_report = st.button("Get Report", disabled=st.session_state.disabled)

        if get_report:

            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            fishcrow_bird_report_app(temp_file_path)

    elif count_choice=="Black Billed Magpie":
        st.markdown("## Black Billed Magpie")

        #st.markdown("#### Enter the path to the audio file")
        #audio_file_path = st.text_input("File Path")

        #NEWLY ADDED STUFF
        st.markdown("### Upload an audio file")
        uploaded_file = st.file_uploader("", type=["mp3", "wav", "ogg"])
        #NEWLY ADDED STUFF ENDS HERE

        detect_button = st.button("Detect")

        st.session_state.disabled = True

        if detect_button and uploaded_file: #change uploaded_file to audio_file_path
            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            calls= count_magpie_birds(temp_file_path) #change it to audio_file_path
            st.markdown(
                f"""
                <div style="text-align: center;">
                    <span style="font-size: 1.5rem; font-weight: bold;">
                        The number of Black Billed Magpie Bird calls detected is:
                    </span>
                    <br>
                    <div style="width: 80px; height: 80px; background-color: #B5C18E; border-radius: 50%; display: flex; justify-content: center; align-items: center; margin: 0 auto;">
                        <span style="font-size: 1.5rem;">{calls}</span>
                    </div>
                    <br>
                </div>
                """,
                unsafe_allow_html=True
            )
            st.session_state.disabled = False

        get_report = st.button("Get Report", disabled=st.session_state.disabled)

        if get_report:
            #NEWLY ADDED STUFF
            st.audio(uploaded_file, format='audio/' + uploaded_file.name.split('.')[-1])

            with tempfile.NamedTemporaryFile(delete=False) as temp_file:
                temp_file_path = temp_file.name
                temp_file.write(uploaded_file.read())
            #NEWLY ADDED STUFF ENDS HERE

            magpie_bird_report_app(temp_file_path)

    else:
        st.subheader("Observing, Documenting, and Recognizing! Dive into the world of birds, discover their profound impact on the natural world.")
        uploaded_file = None

if __name__ == "__main__":
    main()
```
